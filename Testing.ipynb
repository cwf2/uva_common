```python
from dices_common import Text
import os
```

```python
text = Text(
    urn = "urn:cts:greekLit:tlg2046.tlg001.perseus-grc2:1",
    author = "Quintus",
    title = "Posthomerica",
    prefix = "1",
)
```

```python
text.tokenize()
```

```python
tokens = text.get_tokens()
tokens.to_csv(os.path.join("data", "tokens.csv"))
```
